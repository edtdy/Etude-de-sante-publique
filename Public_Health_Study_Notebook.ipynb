```python
bilan_alimentaire = pd.merge(dispo_alimentaire, population, on = 'Zone')
bilan_alimentaire.columns
```

```python
bilan_alimentaire 
```

```python
bilan_alimentaire['Disponibilité alimentaire (Kcal/an)'] = bilan_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)'] \
* 365 * bilan_alimentaire['Valeur'] 
bilan_alimentaire['Disponibilité intérieur (KG)'] = bilan_alimentaire['Disponibilité intérieure']*1000000
```

```python
bilan_alimentaire['Disponibilité de protéines en quantité (Kg/an)']= bilan_alimentaire['Disponibilité de protéines en quantité (g/personne/jour)']\
* 365 * .001 * bilan_alimentaire['Valeur']

bilan_alimentaire['Disponibilité alimentaire en quantité (Kg/an)'] = bilan_alimentaire['Disponibilité alimentaire en quantité (kg/personne/an)'] \
* bilan_alimentaire['Valeur']
```

```python
bilan_alimentaire['Ratio énergie/poids (Kcal/Kg)'] = bilan_alimentaire['Disponibilité alimentaire (Kcal/an)'] \
/ bilan_alimentaire['Disponibilité alimentaire en quantité (Kg/an)']
bilan_alimentaire = bilan_alimentaire.replace([np.inf, -np.inf], np.nan).dropna(subset=["Ratio énergie/poids (Kcal/Kg)"], how="all")
```
